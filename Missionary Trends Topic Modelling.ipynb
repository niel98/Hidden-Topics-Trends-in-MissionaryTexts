```python
import pandas as pd
import pickle
```

```python
data = pd.read_pickle('Mission_dtm.pkl')
data
```

```python
#import the packages
from gensim import matutils, models
import scipy.sparse
```

```python
#Use a term document matrix instead of a document term matrix.
tdm = data.transpose()
tdm.head()
```

```python
#convert the tdm into a sparse matrix cos that is the gensim format
sparse_counts = scipy.sparse.csr_matrix(tdm)
```

```python
#convert the sparse matrix into a gensim corpus
corpus = matutils.Sparse2Corpus(sparse_counts)
```

```python
#Gensim requires a dictionary of all the terms and their respective locations in the term matrix
cv = pickle.load(open('cv_stop.pkl', 'rb'))
id2word = dict((v, k) for k, v in cv.vocabulary_.items())
```

```python
#Now that we have the corpus and the sparse matrix, we proceed to use the lda to find hidden words in the corpus
#we start with k i.e the number of topics to be 2
lda = models.LdaModel(corpus=corpus, id2word=id2word, num_topics=2, passes=10)
lda.print_topics()
```

```python
lda = models.LdaModel(corpus=corpus, id2word=id2word, num_topics=3, passes=10)
lda.print_topics()
```

```python
lda = models.LdaModel(corpus=corpus, id2word=id2word, num_topics=4, passes=10)
lda.print_topics()
```

```python
#These topics aren't really looking good and basically make no sense. we have to tune our terms
#Create a function to pull out nouns and adjectives from the corpus
from nltk import word_tokenize, pos_tag

def nouns(text):
    is_noun = lambda pos: pos[:2] == 'NN' or pos[:2] == 'JJ'
    tokenized = word_tokenize(text)
    all_nouns = [word for (word, pos) in pos_tag(tokenized) if is_noun(pos)]
    return ' '.join(all_nouns)
```

```python
#read in the cleaned data 
data_clean = pd.read_pickle('data_clean.pkl')
data_clean.head()
```

```python
#apply the nouns function to the body to filter only nouns
data_nouns = pd.DataFrame(data_clean.Body_update.apply(nouns))
data_nouns[:2]
```

```python
#Create a new dtm using only nouns
from sklearn.feature_extraction.text import CountVectorizer

cvn = CountVectorizer(stop_words='english')
data_cvn = cvn.fit_transform(data_nouns.Body_update)
data_dtmn = pd.DataFrame(data_cvn.toarray(), columns=cvn.get_feature_names())
data_dtmn.index = data_nouns.index
data_dtmn
```

```python
#create the gensim corpus
corpusn = matutils.Sparse2Corpus(scipy.sparse.csc_matrix(data_dtmn.transpose()))
#create the vocabulary dictionary
idwordn = dict((v, k) for k, v in cvn.vocabulary_.items())
```

```python
ldan = models.LdaModel(corpus=corpusn, id2word=idwordn, num_topics=2, passes=10)
ldan.print_topics()
```

```python
ldan = models.LdaModel(corpus=corpusn, id2word=idwordn, num_topics=3, passes=10)
ldan.print_topics()
```

```python
ldan = models.LdaModel(corpus=corpusn, id2word=idwordn, num_topics=4, passes=10)
ldan.print_topics()
```

```python
#lets increase the number of iterations and plot our final model.
ldan = models.LdaModel(corpus=corpusn, id2word=idwordn, num_topics=4, passes=100)
ldan.print_topics()
```

```python
# the hidden topics found after so many iterations from our final lda model are:
# 0: God, Karen, Mission, Missionary, life
# 1: livingstone, China, great, Lake
# 2: Lord, little, time, work
# 3: Church, years, Bishop
```
